```
#Importações: 
import pandas as pd
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```
#CONSTANTES: 
Graph_Name = input('Indique o nome do gráfico:\n')
```

```
df = pd.read_csv(input('Nome do Arquivo'))
```

```
df.reset_index()
pd.to_datetime(df['etto'])
df.head()
```

```
df['etto']
```

```
df1_convertido_para_float = df[['Temperature','Humidity']]
df1_convertido_para_float = df1_convertido_para_float.apply(pd.to_numeric)
df1_convertido_para_float.plot()
```

```
df2_convertido_para_float = df[['Molhamento','T_DPM']]
df2_convertido_para_float = df2_convertido_para_float.apply(pd.to_numeric)
df2_convertido_para_float.hist()
```

```
df3_convertido_para_float = df['etto']
df3_convertido_para_float.hist()
```

Trying to create the graph


```
fig, ax = plt.subplots(figsize=(10, 6))
plt.title(Graph_Name)
ax.set_xlabel('Dias')
plt.xticks(rotation=45)#Rotaciona os dados do eixo
ax.set_ylabel('Temperatura(°C)')
ax.plot(df1_convertido_para_float['Temperature'],color='#db5f1d', linewidth = 2)
ax2 = ax.twinx() #duplica o eixo
ax2.set_ylabel('Umidade(%)')
ax2.plot(df1_convertido_para_float['Humidity'],color='#4983f5', linewidth = 2)
ax3 = ax.twinx()
ax3.bar(df['etto'], df2_convertido_para_float['Molhamento'],color='#010003', alpha = 0.5)#Alpha da a trnsparência no gráfico
ax3.set_yticklabels([])#apaga os valores que apareceriam no eixo y, lado direito, relacionado ao molhamento
```
