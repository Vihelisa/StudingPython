```
import pandas as pd
df = pd.read_csv(input('Nome do Arquivo'))
df.head()
```

```
df1_convertido_para_float = df[['Temperature','Humidity']]
df1_convertido_para_float = df1_convertido_para_float.apply(pd.to_numeric)
df1_convertido_para_float.plot()
```

```
df2_convertido_para_float = df[['Molhamento','T_DPM']]
df2_convertido_para_float = df2_convertido_para_float.apply(pd.to_numeric)
df2_convertido_para_float.hist()
```

```
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(8, 5))
plt.title('Graph')
#plt.xlabel('Tempo')
ax.plot(df1_convertido_para_float['Humidity'],color='#4983f5', linewidth = 2)
blue_pach = mpatches.Patch(color='#4983f5', label='Humidity(%)')
first_legend = plt.legend(handles=[blue_pach], loc='lower left')
plt.gca().add_artist(first_legend)
orange_pach = mpatches.Patch(color='#db5f1d', label='Temperature(°C)')
plt.legend(handles=[orange_pach], loc='lower right')
ax2 = ax.twinx()
ax2.plot(df1_convertido_para_float['Temperature'],color='#db5f1d', linewidth = 2)

#Histogramas
fig, axs = plt.subplots(2,1, figsize=(7, 4))

axs[0].hist(df2_convertido_para_float['Molhamento'], color='#6096a1')
axs[0].set_ylabel('Molhamento')
axs[1].hist(df2_convertido_para_float['T_DPM'], color='#c46499')
axs[1].set_ylabel('T_DPM')


#plt.show()
ax = fig.add_subplot(axs[0])
ax = fig.add_subplot(axs[1])

plt.show()
```
